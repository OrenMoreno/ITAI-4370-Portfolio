```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
frequency1 = 2400 # MHz (2.4 GHz WiFi)
frequency2 = 5000 # MHz (5 GHz WiFi)
distances = np.linspace(0.1, 20, 100) # km
```

```python
# Free-Space Path Loss (FSPL) calculation
fspl1 = 20*np.log10(distances) + 20*np.log10(frequency1) + 32.44 # FSPL for 2.4 GHz
fspl2 = 20*np.log10(distances) + 20*np.log10(frequency2) + 32.44 # FSPL for 5 GHz
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(distances, fspl1, label="FSPL at 2.4 GHz") # Plot line for 2.4 GHz FSPL
plt.plot(distances, fspl2, label="FSPL at 5 GHz") # Added line for 5 GHz FSPL
plt.xlabel("Distance (km)")
plt.ylabel("Path Loss (dB)")
plt.title("Free-Space Path Loss vs Distance")
plt.grid(True)
plt.legend()
plt.show()
```
